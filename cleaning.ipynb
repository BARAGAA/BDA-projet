## Creating the date table

### creating lables for the months

```python
def make_label(row):
    if row["month"] == 1:
        return "Jan"
    elif row["month"] == 2:
        return "Feb"
    elif row["month"] == 3:
        return "Mar"
    elif row["month"] == 4:
        return "Apr"
    elif row["month"] == 5:
        return "Mai"
    elif row["month"] == 6:
        return "Jun"
    elif row["month"] == 7:
        return "Jul"
    elif row["month"] == 8:
        return "Aug"
    elif row["month"] == 9:
        return "Sep"
    elif row["month"] == 10:
        return "Oct"
    elif row["month"] == 11:
        return "Nov"
    elif row["month"] == 12:
        return "Dec"
    else:
        return "NULL"
        
```

```python
 
def getDay(row):
    date = pd.to_datetime(row["Release"], format = "%Y/%m/%d")
# fetching the day
    return date.day_name()
```

```python
everyNoise = pd.read_csv("CSVs/every-noise-at-once.csv")
everyNoise = everyNoise.dropna(subset=['Release'])
everyNoise = everyNoise.drop_duplicates(subset=['Release'])
everyNoise['month'] = (pd.DatetimeIndex(everyNoise['Release']).month)
everyNoise['year'] = (pd.DatetimeIndex(everyNoise['Release']).year)
everyNoise['day'] = (pd.DatetimeIndex(everyNoise['Release']).day)
everyNoise.day = everyNoise.day.astype(int)
everyNoise.month = everyNoise.month.astype(int)
everyNoise.year = everyNoise.year.astype(int)
date = everyNoise[["Release","day","month","year"]]
date['label'] = date.apply(lambda row: make_label(row),axis=1)
date['week_day'] = date.apply(lambda row: getDay(row),axis=1)

date  = date.drop_duplicates(subset=["Release"])

```

```python
date = date.drop_duplicates()
date.rename(columns={"label":"month_label"},inplace=True)

date
```

```python
f = open("dates.sql", "a",encoding='utf-8')
for index,row in date.iterrows(): 
    f.write("INSERT INTO music_schema.date (" +"date_id"+","+"day_number"+","+"month_number"+","+"year_number"+","+"month_label"+","+"week_day"+")")
    f.write('\n')
    f.write("VALUES(" + "'"+str(row["Release"])+"'"+","+str(row["day"])+","+str(row["month"])+","+str(row["year"])+","+"'"+str(row["month_label"])+"'"+","+"'"+str(row["week_day"])+"');")
    f.write('\n')
    f.write('\n')
f.close()
```

### Creating artist csv

```python
artistsFile = pd.read_csv("CSVs/artists.csv")
for index,row in artistsFile.iterrows(): 
    if not(str(row["artist_id"])) : 
        artistsFile.drop(index,inplace=True)
for index,row in artistsFile.iterrows(): 

    if (str(row["artist_name"])).isnumeric() : 
        artistsFile.drop(index,inplace=True)
for index,row in artistsFile.iterrows(): 

    if ((str(row["followers"]).isnumeric())) == False :
        artistsFile.drop(index,inplace=True)    
artistsFile.to_csv("artists.csv")
```
